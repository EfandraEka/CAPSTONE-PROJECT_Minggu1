# Preprocessing Data

```python
# Mengecek apakah ada missing values di setiap kolom
print(" Missing Values:")
print(df.isnull().sum())
```

Penjelasan: Mengecek jumlah nilai kosong (NaN/null) di setiap kolom. Penting untuk menentukan apakah kita perlu imputasi atau pembersihan data.

```python
# Mengecek dan menghapus data duplikat
print("\n Duplikasi:")
print(f"Jumlah duplikasi: {df.duplicated().sum()}")

# Menghapus baris duplikat jika ada
df = df.drop_duplicates()
```

Penjelasan:

df.duplicated() mengidentifikasi baris yang identik.

drop_duplicates() menghapus baris duplikat untuk menghindari bias pada model.

```python
# Fungsi untuk menghapus outlier menggunakan metode IQR
def remove_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[col] >= lower) & (data[col] <= upper)]
```

Penjelasan:
Fungsi ini menghilangkan outlier berdasarkan metode IQR (interquartile range). Nilai di luar rentang [Q1 - 1.5*IQR, Q3 + 1.5*IQR] dianggap outlier.

```python
# Terapkan fungsi outlier removal ke semua kolom numerik
for col in numerical_cols:
    df = remove_outliers_iqr(df, col)
```

Penjelasan:
Membersihkan outlier dari setiap kolom numerik agar data lebih bersih dan model tidak bias terhadap nilai ekstrem.

```python
from sklearn.preprocessing import LabelEncoder
```

Penjelasan:
Impor LabelEncoder dari Scikit-Learn untuk mengubah data kategorikal menjadi bentuk angka.

```python
# Salin dataframe agar dataset asli tetap aman
df_encoded = df.copy()
```

Penjelasan:
Membuat salinan data agar preprocessing tidak merusak data asli.



```python
# Tentukan kolom-kolom kategorikal yang akan diencoding
categorical_cols = ['Gender', 'family_history_with_overweight', 'FAVC',
                    'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']
```

Penjelasan:
Daftar fitur yang bertipe kategorikal dan perlu diubah menjadi angka agar dapat digunakan dalam model.

```python
# Encode tiap kolom kategorikal
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le  # Simpan encoder untuk digunakan kembali nanti (jika perlu decoding)
```

Penjelasan:
Mengubah data teks (seperti "Yes", "No", "Male", "Female") menjadi angka (0, 1, 2, dst). Encoder disimpan agar bisa dibalik saat interpretasi hasil.

```python
from imblearn.over_sampling import SMOTE
```

Penjelasan:
Mengimpor metode SMOTE dari imbalanced-learn untuk menangani ketidakseimbangan jumlah data pada kelas target.

```python
# Pisahkan fitur dan label
X = df_encoded.drop('NObeyesdad', axis=1)
y = df_encoded['NObeyesdad']
```

 Penjelasan:
X berisi semua fitur input, y adalah target/output yang ingin diprediksi (kategori obesitas).

```python
# Encode label target ke angka
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)
```

 Penjelasan:
Mengubah target multiklas (nama kelas obesitas) menjadi angka agar bisa digunakan dalam model machine learning.



```python
# Lakukan oversampling dengan SMOTE untuk mengatasi ketidakseimbangan kelas
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y_encoded)
```

Penjelasan:
SMOTE menciptakan contoh baru untuk kelas minoritas agar distribusi kelas menjadi seimbang. Ini membantu mencegah model bias ke kelas mayoritas.



```python
from sklearn.preprocessing import StandardScaler

# Lakukan standardisasi ke semua fitur (mean = 0, std = 1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_balanced)
```

Penjelasan:
Standardisasi mengubah skala semua fitur numerik agar memiliki rata-rata 0 dan standar deviasi 1. Ini penting untuk model seperti KNN, SVM, dll.

Kesimpulan :

Pada tahap preprocessing, dataset telah diperiksa dan ditemukan tidak mengandung nilai kosong (missing), serta sebagian kecil duplikasi yang telah dihapus. Outlier pada beberapa fitur numerik juga dibersihkan dengan metode IQR. Semua fitur kategorikal diubah menjadi bentuk numerik menggunakan Label Encoding, termasuk target variabel (NObeyesdad). Untuk menangani ketidakseimbangan kelas target, digunakan metode SMOTE sehingga seluruh kelas obesitas terdistribusi merata. Selanjutnya, dilakukan standardisasi agar seluruh fitur numerik berada dalam skala yang seragam. Dengan demikian, dataset kini telah bersih, seimbang, dan siap digunakan dalam tahap pemodelan machine learning.



# Pemodelan dan Evaluasi

```python
# Import model klasifikasi
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# Import tools evaluasi
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

# Untuk visualisasi
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Penjelasan:

train_test_split: Membagi data menjadi set pelatihan dan pengujian.

RandomForestClassifier, DecisionTreeClassifier, KNeighborsClassifier: Tiga model klasifikasi yang digunakan.

classification_report, accuracy_score, dll.: Menghitung metrik evaluasi.

matplotlib & seaborn: Untuk visualisasi hasil.



```python
# Membagi dataset menjadi data latih dan data uji (80:20)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_balanced, test_size=0.2, stratify=y_balanced, random_state=42
)
```

Penjelasan:

X_scaled: Fitur yang telah dinormalisasi.

y_balanced: Label (tingkat obesitas) yang sudah seimbang (misalnya dengan SMOTE).

test_size=0.2: 20% data untuk pengujian.

stratify=y_balanced: Menjaga proporsi kelas tetap seimbang antara training dan testing.

```python
# Membuat dictionary untuk menyimpan model
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state=42)
}

# Dictionary untuk menyimpan hasil evaluasi tiap model
results = {}
```

Penjelasan:

Kita membuat dictionary models untuk menyimpan 3 algoritma yang ingin kita latih.

results akan menyimpan skor evaluasi dari masing-masing model.

```python
# Loop untuk setiap model
for name, model in models.items():
    # Latih model pada data training
    model.fit(X_train, y_train)

    # Lakukan prediksi pada data testing
    y_pred = model.predict(X_test)

    # Hitung metrik evaluasi
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted')
    rec = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Simpan hasil dalam dictionary
    results[name] = {
        'Model': model,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1 Score': f1,
        'Confusion Matrix': confusion_matrix(y_test, y_pred),
        'Classification Report': classification_report(y_test, y_pred, output_dict=True)
    }
```

Penjelasan:

.fit() digunakan untuk melatih model

.predict() menghasilkan label prediksi pada X_test

Skor dihitung dan disimpan agar bisa dibandingkan antar model

```python
# Loop untuk menampilkan confusion matrix setiap model
for name in results:
    plt.figure(figsize=(6, 4))
    sns.heatmap(results[name]['Confusion Matrix'], annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()
```

Penjelasan:

Confusion Matrix membantu melihat kesalahan prediksi di tiap kelas.

annot=True menunjukkan angka di dalam kotak.



```python
# Ubah hasil metrik ke dalam dataframe agar mudah divisualisasi
metrics_df = pd.DataFrame({
    model: {
        'Accuracy': results[model]['Accuracy'],
        'Precision': results[model]['Precision'],
        'Recall': results[model]['Recall'],
        'F1 Score': results[model]['F1 Score']
    } for model in results
}).T

# Buat plot perbandingan metrik
metrics_df.plot(kind='bar', figsize=(10, 6), colormap='Pastel1')
plt.title("Perbandingan Performa Model Klasifikasi")
plt.ylabel("Skor")
plt.xticks(rotation=0)
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

 Penjelasan:

Menggunakan bar chart untuk membandingkan akurasi, presisi, recall, dan F1-score.


Metrik di atas 0.8 menunjukkan performa baik.

Kesimpulan :

Berdasarkan evaluasi dari tiga model klasifikasi, yaitu Random Forest, K-Nearest Neighbors, dan Decision Tree, dapat disimpulkan bahwa Random Forest memiliki performa terbaik secara konsisten di seluruh metrik evaluasi (akurat, presisi, recall, dan F1). Decision Tree menyusul dengan performa cukup baik, tetapi cenderung overfitting. KNN bekerja baik pada data seimbang, namun lebih sensitif terhadap skala dan distribusi data. Maka, Random Forest menjadi model terbaik untuk digunakan dalam memprediksi tingkat obesitas berdasarkan atribut gaya hidup dan kebiasaan makan.

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

Penjelasan:

Mengimpor model dan GridSearchCV dari scikit-learn:

RandomForestClassifier: Model ensemble berbasis pohon.

KNeighborsClassifier: Model berdasarkan jarak tetangga.

DecisionTreeClassifier: Model pohon keputusan.

```python
def tune_model(model, param_grid, X_train, y_train):
    """
    Fungsi untuk melakukan tuning hyperparameter menggunakan GridSearchCV.
    Mengembalikan model terbaik dan parameter terbaiknya.
    """
    grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3,
                        scoring='accuracy', n_jobs=-1, verbose=1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_, grid.best_params_
```

Penjelasan:

GridSearchCV: menguji semua kombinasi parameter berdasarkan akurasi dengan 3-fold cross-validation.

n_jobs=-1: menggunakan seluruh core CPU agar cepat.

verbose=1: menampilkan progres saat proses tuning berjalan.

grid.best_estimator_: model terbaik berdasarkan hasil tuning.

grid.best_params_: parameter terbaik dari kombinasi yang dicoba.



```python
# Parameter Random Forest
param_rf = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
}

# Parameter KNN
param_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance']
}

# Parameter Decision Tree
param_dt = {
    'max_depth': [None, 10, 20],
    'criterion': ['gini', 'entropy']
}
```

Penjelasan:

Menyusun grid parameter yang akan diuji satu per satu untuk masing-masing model.

Kombinasi ini akan digunakan oleh GridSearchCV.

```python
# Random Forest Tuning
best_rf, best_rf_params = tune_model(RandomForestClassifier(random_state=42), param_rf, X_train, y_train)

# KNN Tuning
best_knn, best_knn_params = tune_model(KNeighborsClassifier(), param_knn, X_train, y_train)

# Decision Tree Tuning
best_dt, best_dt_params = tune_model(DecisionTreeClassifier(random_state=42), param_dt, X_train, y_train)
```

 Penjelasan:

Memanggil fungsi tune_model untuk masing-masing model.

best_rf, best_knn, best_dt: model hasil tuning.

best_rf_params, dll: parameter terbaik yang ditemukan GridSearch.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Simpan model terbaik
tuned_models = {
    'Random Forest (Tuned)': best_rf,
    'KNN (Tuned)': best_knn,
    'Decision Tree (Tuned)': best_dt
}

# Simpan hasil evaluasi
tuned_results = {}

# Evaluasi model
for name, model in tuned_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    tuned_results[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, average='weighted'),
        'Recall': recall_score(y_test, y_pred, average='weighted'),
        'F1 Score': f1_score(y_test, y_pred, average='weighted'),
        'Confusion Matrix': confusion_matrix(y_test, y_pred)
    }
```

Penjelasan:

Melatih ulang model hasil tuning pada data training (model.fit()).

Evaluasi hasil prediksi pada data uji (y_pred) menggunakan metrik akurasi, precision, recall, dan F1 Score.

Disimpan dalam dictionary tuned_results.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Inisialisasi dataframe hasil evaluasi
combined_results = pd.DataFrame()

# Loop setiap model dan gabungkan hasil evaluasi sebelum & sesudah tuning
for name in results.keys():
    tuned_name = f"{name} (Tuned)"

    # Data sebelum tuning
    original_metrics = results[name]
    df_original = pd.DataFrame({
        'Model': name,
        'Tuning': 'Before',
        'Accuracy': original_metrics['Accuracy'],
        'Precision': original_metrics['Precision'],
        'Recall': original_metrics['Recall'],
        'F1 Score': original_metrics['F1 Score']
    }, index=[0])
    combined_results = pd.concat([combined_results, df_original], ignore_index=True)

    # Data sesudah tuning
    if tuned_name in tuned_results:
        tuned_metrics = tuned_results[tuned_name]
        df_tuned = pd.DataFrame({
            'Model': name,
            'Tuning': 'After',
            'Accuracy': tuned_metrics['Accuracy'],
            'Precision': tuned_metrics['Precision'],
            'Recall': tuned_metrics['Recall'],
            'F1 Score': tuned_metrics['F1 Score']
        }, index=[0])
        combined_results = pd.concat([combined_results, df_tuned], ignore_index=True)
    else:
        print(f" Warning: Tidak ditemukan hasil tuning untuk model '{name}'.")

# Tampilkan hasil evaluasi lengkap
print("\n Perbandingan Performa Setiap Model (Sebelum & Sesudah Tuning):\n")
print(combined_results)

# Visualisasi: Subplot untuk setiap metrik
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# Loop setiap metrik untuk visualisasi
for idx, metric in enumerate(metrics):
    ax = axes[idx]
    for model in combined_results['Model'].unique():
        subset = combined_results[combined_results['Model'] == model]
        bars = ax.bar(
            x=[f"{model} ({row['Tuning']})" for _, row in subset.iterrows()],
            height=subset[metric],
            label=model
        )
    ax.set_title(metric)
    ax.set_ylim(0, 1.05)
    ax.set_ylabel("Score")
    ax.tick_params(axis='x', rotation=45)
    ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.suptitle(" Perbandingan Performa Model Sebelum dan Sesudah Tuning", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

Melakukan iterasi untuk setiap model (Random Forest, KNN, Decision Tree), baik sebelum maupun sesudah tuning.

Membuat dua baris dataframe:

- After: data sesudah tuning.

Keduanya disatukan dalam combined_results.

Random Forest menunjukkan performa terbaik, terutama dalam akurasi dan F1 Score.
KNN dan Decision Tree juga membaik setelah tuning jumlah tetangga dan kedalaman pohon.
